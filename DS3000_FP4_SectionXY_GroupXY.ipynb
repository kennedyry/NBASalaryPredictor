## 1. INTRODUCTION

In the NBA, it is important that players are performing well relative to their contract and that general managers of teams give players a contract that correlates to the player's performance. Our project aims to determine an NBA player's salary relative to their performance and if their current contract value is reflected in a player's stats. We want to figure out what a player is worth based on their playing statistics.

In the NBA, especially with a salary cap, it is important to utilize the money that you give to players wisely. In order to win the championship, teams must not give out large contracts to players that do not actually deserve it. You must have players that perform at or above their contract value, in order to be successful in the league. The insights found from this project could help NBA general managers determine a player's worth when negotiating contracts and show teams which players should be traded due to their performance relative to the value of their contract. 

**Previous Work**

There has been quite a bit of research done in the past, relating to predicting the salary of an NBA player based on their stats. This is because there is such a large amount of numerical data, that is very easily attainable. The NBA is also a very popular league, so many enjoy working with data related to a subject that they are interested in.

In one previous study (Ando, 2018), they used a Linear Regression Model in R to create a model that would predict salary in the 17-18 season based on a player's stats from the 16-17 season. In another study (Rosson 2019), they used NBA player data from 1995 to 2017, along with salary cap information and team wins per season. Since total salary varies from year to year, they normalized player salary by creating a percentage of salary cap field for each player. They used a heat map to find what fields were most correlated with the target value. They used a Linear Regression model for their data, where they found that the most undervalued players were good players who were still on their rookie contract. Another study (Wu 2018), used NBA player data from 2014-2017 and cross-referenced with a list of free agents for each of the seasons. They also standardized salaries by dividing by the total cap for that season, but they went a step further and grouped the percentage of the cap into 8 clusters. They used a kNN classification model. The last related study (Xu, 2019), used NBA data from 1990 to 2018. Similar to other previous studies, the total salary was standardized, as to be able to compare across years. They compared the ability of three different kinds of regression models: Linear, Decision Tree, and Random Forest. They compared the MAE (mean absolute error), MSE (mean squared error), and RMSE (root mean squared error) of each model, and found that the Linear Regression model worked the best for their data.

## 2. METHOD

### 2.1. Data Acquisition

We obtained our data from two places. We acquired our player statistics from the website Basketball Reference (https://www.basketball-reference.com/), where we went through grabbing the player statistics for years of interest and then combined all of this data into an overall data file. We then grabbed the player salaries from a website called Celtics Hub (https://www.celticshub.com/2017/12/07/nba-player-salaries-1991-2017/), which included the salaries from 1991-2017 for all players which we were then able to merge into our main data set to include these as our target variable. Our project's dataset consists of 1442 rows and 53 columns lisiting all players in the NBA from 2013-2015 along with their recorded statistics and their contract value. Every player's contract value is represented by a salary cap percentage, which represents how much of the team's total salary goes to the player. The features of this dataset are a player's stats and the target value is the salary cap percentage. We will then use this data to predict the value of any given player's contract. The link to our cleaned data set is found here https://github.com/kennedyry/DS3000Project/blob/main/data/data.csv.

### 2.2. Data Analysis

For our hypotheses, the independent variable will be the features of our dataset, namely the player's season statistics between 2013 and 2015. The dependent variable will be the percentage of a team's cap that the model predicts a player is worth. The outcome variable will be this cap percentage and will be derived using the highest weighing features after running an algorithm like model based feature selection. Since we generally do not know which statistics of a basketball player are more valuable, we figured letting the model determine the weights would be more effective. The problem we will try to tackle is a supervised ML problem and more specifically a problem of regression since we already have the 'correct' value of a player in their salary (in terms of cap percentage) of a given year and since the output variable will ultimately be a value on a continuous scale. We plan on using the regression form of KNN along with a lasso regressor and linear regression to model our data. The reason we chose KNN is because we believe that players with similar contracts tend to have similar playing statistics and thus nearest neighbors algorithm in KNN would be really useful. We chose to try linear regression because our background research suggested that linear regression would be effective for our task. We chose to use the Lasso regressor as well because our data contains a lot of features which may not matter heavily on the calculation. We believe that only a few will matter in the long run and thus using L1 regularization will be useful.

```python
# Imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeRegressor
```

```python
# Data Cleaning:
# We have previously cleaned the data and exported to a csv file and thus do not need to do this again. Below
# is a copy of the code that we ran to clean our data.
# We will not run this code ourselves in this report. If you want to run it yourself, feel free.
# Below is DataMerge.ipynb

df = pd.DataFrame()
duplicates = ['Rk','Pos','Age','Tm','G','MP']
    
for year in ['2013','2014','2015']:
    stat = pd.read_csv("data/{}stats.csv".format(year))
    advanced = pd.read_csv("data/{}advancedstats.csv".format(year))
    advanced = advanced.drop(columns = duplicates)
    stat = stat.merge(advanced, on = 'Player')
    stat['Year'] = int(year)
    #print(advanced.keys())
    #print('Stat', stat.keys())
    stat = stat.drop_duplicates(subset = 'Player', keep = 'first')
    df = df.append(stat)
    #print(df.keys())
    #print('Stat', stat.keys())

df['UID'] = df['Player'].apply(lambda name : name.split('\\')[1])
df['Player'] = df['Player'].apply(lambda name : name.split('\\')[0])
df = df.drop(columns = 'Rk')
   
df = df.drop(columns = ['Unnamed: 19','Unnamed: 24'])
df = df.fillna(0)

cd = pd.read_csv('data/NBA_Salary_History - HistoryFormatted.csv')
cd = cd.drop(columns = ['BelowMin', 'Unnamed: 5', 'Team'])
cd = cd.rename(columns = {'YearEnd':'Year', ' Salary ':'Salary'})


cd['Salary'] = cd['Salary'].apply(lambda x : x.replace(',',''))

cd.head()


df = df.merge(cd, how = 'inner', on=['Year','Player'])

capD = pd.read_csv('data/capdata.csv')
capD = capD.set_index('Year')

df['Cap%'] = df.T.apply(lambda row : int(row['Salary']) / capD.loc[row['Year']]['Salary Cap'])


df.head()

df.to_csv('data/data.csv')
```

```python
# Data Wrangling:
df = pd.read_csv('data/data.csv', index_col=0)
position_dictionary = { 'G': 1, 'SG': 2, 'SF': 3, 'PF': 4, 'C': 5 }
index = 6
for row in set(df['Pos']):
    if not row in position_dictionary:
        position_dictionary[row] = index
        index += 1
df['Pos'] = df['Pos'].apply(lambda pos: position_dictionary[pos])

features = df.drop(columns = ['Cap%', 'UID', 'Player', 'Tm', 'Salary'], axis=1)
target = df['Cap%']

df.head()
```

```python
# Data Preprocessing:
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=3000)
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Feature Extraction: 
select = SelectFromModel(DecisionTreeRegressor(random_state=3000), threshold = 'median')

select.fit(X_train, y_train)
X_train_selected = select.transform(X_train)
X_test_selected = select.transform(X_test)

print(features.columns)
print(select.get_support())
print(select.estimator_.feature_importances_)
```

### 3.3. Model Construction
* Conduct your hypothesis test(s) here.
* For your machine learning question(s), use the Training, Validation, and Testing approach through GridSearch
* Apply machine learning algorithms (apply at least three different algorithms)
* Train your algorithms

```python
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVR
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
estimators = {'Lasso': Lasso(),
              'k-Nearest Neighbor': KNeighborsRegressor(),
              'Linear Regression': LinearRegression()}
param_grid_list = {
    'Lasso': {'alpha': [i for i in np.arange(1, 0, -0.1)]},
    'k-Nearest Neighbor': {'metric': ['euclidean', 'manhattan', 'minkowski'], 'n_neighbors': [1, 5, 10]},
    
}
# Note, looking at the linear regression documentation, it seems like there is not a value to be hypertuned
# and thus it does not exist in our param_grid_list

for param_grid in param_grid_list.keys():
    grid_search = GridSearchCV(estimators[param_grid], param_grid_list[param_grid], cv=5)
    grid_search.fit(X=X_train_selected, y=y_train)
    print(param_grid)
    print("Best parameters: ", grid_search.best_params_)
    print("Training set score with best parameters: ", grid_search.best_score_)
    print("Test set score with best parameters: ", grid_search.score(X_test_selected, y_test))
    print('')
```

```python
updated_estimators = {
    'Lasso': Lasso(alpha = 0.1),
    'k-Nearest Neighbor': KNeighborsRegressor(metric='manhattan', n_neighbors=10),
    'Linear Regression': LinearRegression(),
    'Support Vector Machine': LinearSVR(max_iter=1000000),
    'Ridge': Ridge(),
    'Decision Tree Regressor': DecisionTreeRegressor()
}

for estimator in updated_estimators.keys():
    model = updated_estimators[estimator].fit(X=X_train_selected, y=y_train)
    print(f'{estimator} performance with selected features:')
    print("\tR-squared value for training set: ", r2_score(y_train, model.predict(X_train_selected)))
    print("\tR-squared value for testing set: ", r2_score(y_test, model.predict(X_test_selected)))
    print('')
```
